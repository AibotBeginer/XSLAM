## Newton's Equations of Motion

牛顿运动方程告诉我们，给定一个系统的恒定速度 $v$，我们可以计算它在时间 $t$ 后的位置 $x$：

$$x = vt + x_0$$

例如，如果我们从位置 13 开始，我们的速度是 10 m/s，我们行进 12 秒，我们的最终位置是 133 ($10\times 12 + 13$)。

我们可以将恒定加速度与这个方程结合起来

$$x = \frac{1}{2}at^2 + v_0t + x_0$$

如果我们假设jerk恒定，我们得到

$$x = \frac{1}{6}jt^3 +  \frac{1}{2}a_0 t^2 + v_0 t + x_0$$

这些方程是通过积分微分方程生成的。 给定一个恒定的速度 v 我们可以用方程计算随时间行进的距离

$$x = vt + x_0$$

我们可以得出

$$\begin{aligned} v &= \frac{dx}{dt}\\
dx &= v\, dt \\
\int_{x_0}^x\, dx &= \int_0^t v\, dt\\
x - x_0 &= vt - 0\\
x &= vt + x_0\end{aligned}$$

当您设计卡尔曼滤波器时，您从描述系统动力学的微分方程系统开始。 大多数微分方程系统不容易以这种方式积分。 我们从牛顿方程开始，因为我们可以积分并得到一个封闭形式的解，这使得卡尔曼滤波器更容易设计。 另一个好处是牛顿方程是用于跟踪移动物体的正确方程，这是卡尔曼滤波器的主要用途之一。

     或者预测更准确
     4.在预测和测量之间设置状态
     关于比例因子
     5. 根据我们的确定程度更新对状态的信念
     在测量中

单变量卡尔曼滤波器用单变量高斯表示状态。 自然地，多元卡尔曼滤波器将使用多元高斯来表示状态。 我们在上一章中了解到，多元高斯使用向量作为均值，使用矩阵作为协方差。 这意味着卡尔曼滤波器需要使用线性代数来执行估计。

我不想让你记住这些方程，但我在下面列出了单变量和多变量方程。 它们非常相似。

<u>**Predict**</u>

$\begin{array}{|l|l|l|}
\hline
\text{Univariate} & \text{Univariate} & \text{Multivariate}\\
& \text{(Kalman form)} & \\
\hline
\bar \mu = \mu + \mu_{f_x} & \bar x = x + dx & \bar{\mathbf x} = \mathbf{Fx} + \mathbf{Bu}\\
\bar\sigma^2 = \sigma_x^2 + \sigma_{f_x}^2 & \bar P = P + Q & \bar{\mathbf P} = \mathbf{FPF}^\mathsf T + \mathbf Q \\
\hline
\end{array}$

不用担心线性代数的细节，我们可以看到：

$\mathbf x,\, \mathbf P$ 是状态均值和协方差。 它们对应于 $x$ 和 $\sigma^2$。

$\mathbf F$ 是*状态转换函数*。 当乘以 $\bf x$ 时，它计算先验。

$\mathbf Q$ 是过程协方差。 它对应于 $\sigma^2_{f_x}$。

$\mathbf B$ 和 $\mathbf u$ 对我们来说是新的。 它们让我们对系统的控制输入进行建模。

$\begin{array}{|l|l|l|}
\hline
\text{Univariate} & \text{Univariate} & \text{Multivariate}\\
& \text{(Kalman form)} & \\
\hline
& y = z - \bar x & \mathbf y = \mathbf z - \mathbf{H\bar x} \\
& K = \frac{\bar P}{\bar P+R}&
\mathbf K = \mathbf{\bar{P}H}^\mathsf T (\mathbf{H\bar{P}H}^\mathsf T + \mathbf R)^{-1} \\
\mu=\frac{\bar\sigma^2\, \mu_z + \sigma_z^2 \, \bar\mu} {\bar\sigma^2 + \sigma_z^2} & x = \bar x + Ky & \mathbf x = \bar{\mathbf x} + \mathbf{Ky} \\
\sigma^2 = \frac{\sigma_1^2\sigma_2^2}{\sigma_1^2+\sigma_2^2} & P = (1-K)\bar P &
\mathbf P = (\mathbf I - \mathbf{KH})\mathbf{\bar{P}} \\
\hline
\end{array}$

$\mathbf H$ 是测量函数。 我们还没有在本书中看到这一点，我稍后会解释它。 如果你在脑海中从方程中删除 $\mathbf H$，你应该能够看到这些方程也是相似的。

$\mathbf z,\, \mathbf R$ 是测量均值和噪声协方差。 它们对应于单变量过滤器中的 $z$ 和 $\sigma_z^2$（我已将 $\mu$ 替换为 $x$ 用于单变量方程，以使符号尽可能相似）。

$\mathbf y$ 和 $\mathbf K$ 是残差和卡尔曼增益。

细节将与单变量过滤器不同，因为它们是向量和矩阵，但概念完全相同：

- 使用高斯表示我们对状态和误差的估计
- 使用高斯表示测量及其误差
- 使用高斯表示过程模型
- 使用过程模型预测下一个状态（先验）
- 在测量和先验之间形成估计部分

作为设计师，你的工作将是设计状态 $\left(\mathbf x, \mathbf P\right)$，过程 $\left(\mathbf F, \mathbf Q\right)$，测量 $\left (\mathbf z, \mathbf R\right)$，以及测量函数 $\mathbf H$。 如果系统具有控制输入，例如机器人，您还将设计 $\mathbf B$ 和 $\mathbf u$。

## Tracking a Dog

让我们回到我们尝试过且真实的跟踪狗的问题。 这次我们将包含上一章的基本见解，并使用*隐藏变量*来改进我们的估计。 我可以从数学开始，但是让我们实现一个过滤器，边学习边学习。 从表面上看，数学与前几章不同，可能更复杂，但思想都是一样的——我们只是乘以和加高斯。

我们首先为狗编写一个模拟。 模拟将运行“count”步，每一步将狗向前移动大约 1 米。 在每一步，速度将根据过程方差“process_var”而变化。 更新位置后，我们使用假定的传感器方差“z_var”计算测量值。 该函数返回一个包含位置和另一个测量值的 NumPy 数组。

````python
import math
import numpy as np
from numpy.random import randn

def compute_dog_data(z_var, process_var, count=1, dt=1.):
    "returns track, measurements 1D ndarrays"
    x, vel = 0., 1.
    z_std = math.sqrt(z_var) 
    p_std = math.sqrt(process_var)
    xs, zs = [], []
    for _ in range(count):
        v = vel + (randn() * p_std)
        x += v*dt        
        xs.append(x)
        zs.append(x + randn() * z_std)        
    return np.array(xs), np.array(zs)
````

### Design State Variable

我们之前使用高斯在一个维度上跟踪了一只狗。均值$(\mu)$代表最可能的位置，方差($\sigma^2$)代表该位置的概率分布。位置是系统的*状态*，我们称$\mu$为*状态变量*。

在这个问题中，我们将跟踪狗的位置和速度。这需要我们使用由状态向量 $\mathbf x$ 及其对应的协方差矩阵 $\mathbf P$ 表示的多元高斯。

状态变量可以是*观察变量* - 由传感器直接测量，或*隐藏变量* - 从观察变量推断。对于我们的狗跟踪问题，传感器仅读取位置，因此可以观察位置并隐藏速度。我们将很快学习如何跟踪隐藏变量。

重要的是要了解跟踪位置和速度是一种设计选择，具有我们尚未准备好探索的含义和假设。例如，我们还可以跟踪加速度，甚至是加加速度。现在，回想一下，在上一章中，我们展示了在协方差矩阵中包含速度会导致位置变化小得多。我们将在本章后面学习卡尔曼滤波器如何计算隐藏变量的估计值。

在单变量章节中，我们用标量值表示狗的位置（例如 $\mu=3.27$）。 在上一章中，我们学习了对多个变量使用多元高斯。 例如，如果我们想指定 10.0 m 的位置和 4.5 m/s 的速度，我们会这样写：

$$\mu = \begin{bmatrix}10.0\\4.5\end{bmatrix}$$

卡尔曼滤波器是使用线性代数实现的。 我们使用一个$n\times 1$ 矩阵（称为*vector*）来存储$n$ 个状态变量。 对于狗跟踪问题，我们使用 $x$ 来表示位置，并使用 $x$ 的一阶导数 $\dot x$ 来表示速度。 我使用牛顿的点符号表示导数； $\dot x$ 表示 x 关于 t 的一阶导数：$\dot x = \frac{dx}{dt}$。 卡尔曼滤波器方程使用 $\mathbf x$ 表示状态，因此我们将 $\mathbf x$ 定义为：

$$\mathbf x =\begin{bmatrix}x \\ \dot x\end{bmatrix}$$

我们使用 $\mathbf x$ 而不是 $\mu$，但要认识到这是多元高斯的均值。

另一种写法是 $\mathbf x =\begin{bmatrix}x & \dot x\end{bmatrix}^\mathsf T$ 因为行向量的转置是列向量。 这种表示法更容易在文本中使用，因为它占用的垂直空间更少。

$\mathbf x$ 和位置 $x$ 巧合地同名。 如果我们在 y 轴上跟踪狗，我们会写成 $\mathbf x =\begin{bmatrix}y & \dot y\end{bmatrix}^\mathsf T$，而不是 $\mathbf y =\begin{bmatrix}y & \dot y\end{bmatrix}^\mathsf T$. $\mathbf x$ 是卡尔曼滤波器文献中使用的状态变量的标准名称，我们不会对其进行更改以赋予它更有意义的名称。 这种命名的一致性使我们能够与同行交流。

###  Design the Process Model

下一步是设计*流程模型*。 它是描述系统行为的数学模型。 过滤器使用它来预测离散时间步后的状态。 我们使用一组描述系统动力学的方程来做到这一点。

在单变量章节中，我们用

$$ x = v \Delta t + x_0$$

我们实现如下：

```python
def predict(pos, movement):
    return gaussian(pos.mean + movement.mean, 
                    pos.var + movement.var)
```

我们将在本章中做同样的事情，使用多元高斯而不是单变量高斯。 你可以想象这样的实现：

$$ \mathbf x = \begin{bmatrix}5.4\\4.2\end{bmatrix}, \, \, 
\dot{\mathbf x} =  \begin{bmatrix}1.1\\0.\end{bmatrix} \\
\mathbf x = \dot{\mathbf x}t + \mathbf x$$

但我们需要概括这一点。 卡尔曼滤波器方程适用于任何线性系统，而不仅仅是牛顿系统。 也许您正在过滤的系统是化工厂中的管道系统，给定管道中的流量由不同阀门设置的线性组合决定。

$$\mathtt{pipe_1} = 0.134(\mathtt{valve}_1) + 0.41(\mathtt{valve}_2 - \mathtt{valve}_3) + 1.34$$
$$\mathtt{pipe_2} = 0.210(\mathtt{valve}_2) - 0.62(\mathtt{valve}_1 - \mathtt{valve}_5) + 1.86$$

线性代数有一种强大的方式来表达方程组。 拿这个系统

$$\begin{cases}
2x+3y=8\\4x-y=2
\end{cases}$$

我们可以这样写成矩阵形式：

$$\begin{bmatrix}2& 3 \\ 4&-1\end{bmatrix} \begin{bmatrix}x\\y\end{bmatrix} = \begin{bmatrix}8\\2\end{bmatrix}$$

如果您在此等式中执行 [矩阵乘法](https://en.wikipedia.org/wiki/Matrix_multiplication#General_definition_of_the_matrix_product)，则结果将是上述两个等式。 在线性代数中，我们将其写为 $\mathbf{Ax}=\mathbf B$，其中

$$\mathbf{A} = \begin{bmatrix}2& 3 \\ 4&-1\end{bmatrix},\, \mathbf x = \begin{bmatrix}x\\y\end{bmatrix}, \mathbf B=\begin{bmatrix}8\\2\end{bmatrix}$$

然后我们可以使用 SciPy 的 `linalg` 包来求解 $\mathbf x$：

````python
from scipy.linalg import solve
A = np.array([[2, 3],[4, -1]])
b = np.array([[8], [2]])
x = solve(A, b)
x
````

我们使用过程模型来执行*创新*，因为方程告诉我们下一个状态将被赋予当前状态。 卡尔曼滤波器使用这个线性方程实现这一点，其中 $\mathbf{\bar x}$ 是 *prior* 或预测状态：

$$\mathbf{\bar x} = \mathbf{Fx}$$

我们可以明确表示为

$$\begin{bmatrix} \bar x \\ \dot{\bar x}\end{bmatrix} = \begin{bmatrix}? & ? \\? & ?\end{bmatrix}\begin{bmatrix}x\\\dot x\end{bmatrix}$$

作为卡尔曼滤波器设计者，我们的工作是指定 $\mathbf F$ 使得 $\bar{\mathbf x} = \mathbf{Fx}$ 为我们的系统执行创新（预测）。 为此，我们需要为每个状态变量建立一个方程。 在我们的问题 $\mathbf x = \begin{bmatrix}x & \dot x\end{bmatrix}^\mathtt{T}$，所以我们需要一个方程来计算位置 $x$ 和另一个来计算速度 $ \dot x$ 。 我们已经知道位置创新的方程式：

$$\bar x = x + \dot x \Delta t$$

我们的速度方程是什么？ 对于我们的狗的速度将如何随时间变化，我们没有预测模型。 在这种情况下，我们假设它在创新之间保持不变。 当然这并不完全正确，但只要速度在每次创新中变化不大，您就会发现过滤器的性能非常好。 所以我们说

$$\bar{\dot x} = \dot x$$

这为我们提供了系统的流程模型

$$\begin{cases}
\begin{aligned}
\bar x &= x + \dot x \Delta t \\
\bar{\dot x} &= \dot x
\end{aligned}
\end{cases}$$


这正确地为状态中的每个变量都有一个方程，在左侧隔离。 我们需要用 $\bar{\mathbf x} = \mathbf{Fx}$ 的形式来表达这组方程。 重新排列术语可以更容易地了解要做什么。

$$\begin{cases}
\begin{aligned}
\bar x &= 1x + &\Delta t\, \dot x \\
\bar{\dot x} &=0x + &1\, \dot x
\end{aligned}
\end{cases}$$


我们可以将其重写为矩阵形式

$$\begin{aligned}
\begin{bmatrix}\bar x \\ \bar{\dot x}\end{bmatrix} &= \begin{bmatrix}1&\Delta t  \\ 0&1\end{bmatrix}  \begin{bmatrix}x \\ \dot x\end{bmatrix}\\
\mathbf{\bar x} &= \mathbf{Fx}
\end{aligned}$$

$\mathbf F$ 被称为*状态转移函数*或*状态转移矩阵*。 在后面的章节中，它将是一个真正的函数，而不是一个矩阵，所以称它为一个函数更笼统一些。

````python
dt = 0.1
F = np.array([[1, dt],
              [0, 1]])
F
````

让我们测试一下！ FilterPy 有一个 `predict` 方法，它通过计算 $\mathbf{\bar x} = \mathbf{Fx}$ 来执行预测。 让我们调用它，看看会发生什么。 我们将位置设置为 10.0，速度设置为 4.5 米/秒。 我们已经定义了`dt = 0.1`，这意味着时间步长是0.1秒，所以我们预计创新后的新位置是10.45米。 速度应该不变。

````python
from filterpy.kalman import predict

x = np.array([10.0, 4.5])
P = np.diag([500, 49])
F = np.array([[1, dt], [0, 1]])

# Q is the process noise
x, P = predict(x=x, P=P, F=F, Q=0)
print('x =', x)
````

这行得通。 如果我们连续多次调用`predict()`，则每次都会更新值。

````python
for _ in range(4):
    x, P = predict(x=x, P=P, F=F, Q=0)
    print('x =', x)
````

### Design Process Noise

快速回顾*过程噪音*。 一辆汽车在巡航控制开启的情况下沿着道路行驶； 它应该以恒定的速度行驶。 我们用 $\bar x_k=\dot x_k\Delta t + x_{k-1}$ 对此进行建模。 但是，它受到许多未知因素的影响。 巡航控制不能完美地保持恒定速度。 风会影响汽车，山丘和坑洼也是如此。 乘客摇下车窗，改变汽车的阻力轮廓。

我们可以用微分方程来模拟这个系统

$$\dot{\mathbf x} = f(\mathbf x) + w$$

其中 $f(\mathbf x)$ 为状态转换建模，$w$ 是*白色过程噪声*。

我们将在**卡尔曼滤波器数学**一章中学习如何从一组微分方程到卡尔曼滤波器矩阵。在本章中，我们利用牛顿已经为我们推导出运动方程这一事实。现在你只需要知道我们通过将过程噪声协方差矩阵 $\mathbf Q$ 添加到协方差 $\mathbf P$ 来解释系统中的噪声。我们不向 $\mathbf x$ 添加任何内容，因为噪声是 *white* - 这意味着噪声的平均值将为 0。如果平均值为 0，则 $\mathbf x$ 不会改变。

单变量卡尔曼滤波器使用“方差 = 方差 + 过程噪声”来计算预测步骤方差的方差。多元卡尔曼滤波器的作用相同，本质上是“P = P + Q”。我说“基本上”是因为在协方差方程中还有其他与噪声无关的项，我们稍后会看到。

推导过程噪声矩阵可能非常困难，我们将把它推迟到卡尔曼数学章节。现在知道 $\mathbf Q$ 等于白噪声 $w$ 的期望值，计算为 $\mathbf Q = \mathbb E[\mathbf{ww}^\mathsf T]$。在本章中，我们将专注于建立对修改此矩阵如何改变过滤器行为的直观理解。

FilterPy 为本章的运动学问题提供了计算 $\mathbf Q$ 的函数。 `Q_discrete_white_noise` 有 3 个参数。 `dim`，指定矩阵的维度，`dt`，以秒为单位的时间步长，以及`var`，噪声的方差。简而言之，它在我们稍后讨论的假设下对给定时间段内的噪声进行离散化。此代码计算方差为 2.35 且时间步长为 1 秒的白噪声的 $\mathbf Q$：

````python
from filterpy.common import Q_discrete_white_noise
Q = Q_discrete_white_noise(dim=2, dt=1., var=2.35)
print(Q)
````

### Design the Measurement Function

卡尔曼滤波器计算我们所说的*测量空间*中的更新步骤。 我们在单变量章节中大多忽略了这个问题，因为它增加了复杂性。 我们使用报告他位置的传感器跟踪我们的狗的位置。 计算 *residual* 很容易 - 从测量中减去过滤器的预测位置：

$$ \mathtt{residual} = \mathtt{measured\, \, position} - \mathtt{predicted\, \, position}$$

我们需要计算残差，因为我们通过卡尔曼增益对其进行缩放以获得新的估计。

如果我们尝试使用输出与温度读数相对应的电压的温度计来跟踪温度会发生什么？ 残差计算等式将毫无意义； 你不能从电压中减去温度。

$$ \mathtt{residual} = \mathtt{voltage} - \mathtt{temperature}\;\;\;(NONSENSE!)$$

我们需要将温度转换为电压，以便我们可以执行减法。 对于温度计，我们可以写：

```python
CELSIUS_TO_VOLTS = 0.21475
residual = voltage - (CELSIUS_TO_VOLTS * predicted_temperature)
```

卡尔曼滤波器通过让您提供将状态转换为测量值的*测量函数*来概括此问题。

我们为什么要在测量空间工作？ 为什么不通过将电压转换为温度来在状态空间中工作，让残差成为温度差？

我们不能这样做，因为大多数测量值都不是*可逆的*。 跟踪问题的状态包含隐藏变量 $\dot x$。 无法将位置测量值转换为包含速度的状态。 另一方面，将包含位置和速度的状态转换为仅包含位置的等效“测量”是微不足道的。 我们必须在测量空间中工作以使残差的计算成为可能。

测量 $\mathbf z$ 和状态 $\mathbf x$ 都是向量，因此我们需要使用矩阵来执行转换。 执行此步骤的卡尔曼滤波器方程为：

$$\mathbf y = \mathbf z - \mathbf{H \bar x}$$

其中$\mathbf y$ 是残差，$\mathbf{\bar x}$ 是先验，$\mathbf z$ 是测量值，$\mathbf H$ 是测量函数。 因此，我们采用先验，通过将其乘以 $\mathbf H$ 将其转换为测量值，然后从测量值中减去它。 这给了我们在测量空间中的预测和测量之间的差异！
<img src="./figs/residual_chart_with_h.png">

我们需要设计 $\mathbf H$ 以便 $\mathbf{H\bar x}$ 产生一个测量值。 对于这个问题，我们有一个测量位置的传感器，所以 $\mathbf z$ 将是一个单变量向量：

$$\mathbf z = \begin{bmatrix}z\end{bmatrix}$$

残差方程的形式为

$$
\begin{aligned}
\textbf{y} &= \mathbf z - \mathbf{H\bar x}  \\
\begin{bmatrix}y \end{bmatrix} &= \begin{bmatrix}z\end{bmatrix} - \begin{bmatrix}?&?\end{bmatrix} \begin{bmatrix}x \\ \dot x\end{bmatrix}
\end{aligned}
$$

$\mathbf H$ 必须是 1x2 矩阵，$\mathbf{Hx}$ 必须是 1x1。 回想一下，将矩阵 $m\times n$ 乘以 $n\times p$ 会产生一个 $m\times p$ 矩阵。

我们希望将位置 $x$ 乘以 1 以获得相应的位置测量值。 我们不需要使用速度来找到相应的测量值，因此我们将 $\dot x$ 乘以 0。

$$\begin{aligned}
\textbf{y} &= \mathbf z - \begin{bmatrix}1&0\end{bmatrix} \begin{bmatrix}x \\ \dot x\end{bmatrix} \\
&= [z] - [x]
\end{aligned}$$

因此，对于我们的卡尔曼滤波器，我们设置

$$\mathbf H=\begin{bmatrix}1&0\end{bmatrix}$$

````python
H = np.array([[1., 0.]])
````

我们设计了大部分卡尔曼滤波器。 剩下的就是对传感器中的噪声进行建模。

####  Design the Measurement

测量是用测量平均值 $\mathbf z$ 和测量协方差 $\mathbf R$ 来实现的。

$\mathbf z$ 很简单。 它包含作为向量的测量值。 我们只有一个测量值，所以我们有：

$$\mathbf z = \begin{bmatrix}z\end{bmatrix}$$

如果我们有两个传感器或测量值，我们将拥有：

$$\mathbf z = \begin{bmatrix}z_1 \\ z_2\end{bmatrix}$$

*测量噪声矩阵*将我们传感器中的噪声建模为协方差矩阵。 在实践中，这可能很困难。 一个复杂的系统可能有很多传感器，它们之间的相关性可能不清楚，而且它们的噪声通常不是纯高斯的。 例如，如果温度高，传感器可能会偏向读数高，因此噪声在平均值两侧分布不均。 稍后我们将学习处理这些问题。

卡尔曼滤波器方程对测量噪声使用协方差矩阵 $\mathbf R$。 该矩阵的维度为 $m{\times}m$，其中 $m$ 是传感器的数量。 它是一个协方差矩阵，用于解释传感器之间的相关性。 我们只有 1 个传感器，所以 R 是：

$$R = \begin{bmatrix}\sigma^2_z\end{bmatrix}$$

如果 $\sigma^2_z$ 是 5 米的平方，我们将有 $R = \begin{bmatrix}5\end{bmatrix}$。

如果我们有两个位置传感器，第一个的方差为 5 m$^2$，第二个的方差为 3 m$^2$，我们会写

$$R = \begin{bmatrix}5&0\\0&3\end{bmatrix}$$

我们将方差放在对角线上，因为这是一个*协方差*矩阵，其中方差位于对角线上，而协方差（如果有的话）位于非对角线元素中。 这里我们假设两个传感器之间的噪声没有相关性，因此协方差为 0。

对于我们的问题，我们只有一个传感器，因此我们可以将其实现为

````python
R = np.array([[5.]])
````

## Implementing the Kalman Filter

我已经为您提供了过滤器的所有代码，但现在让我们将它收集在一个地方。 首先我们构造一个 `KalmanFilter` 对象。 我们必须使用 `dim_x` 参数指定状态中的变量数量，并使用 `dim_z` 指定测量次数。 我们在状态中有两个随机变量和一个测量值，所以我们写：

````python
from filterpy.kalman import KalmanFilter
dog_filter = KalmanFilter(dim_x=2, dim_z=1)
print('x = ', dog_filter.x.T)
print('R = ', dog_filter.R)
print('Q = \n', dog_filter.Q)
# etc...
````

现在我们用对我们的问题有效的值初始化过滤器的矩阵和向量。 我把它放在一个函数中，允许您为“R”、“P”和“Q”指定不同的初始值，并将它放在一个辅助函数中。 我们将创建和运行许多这样的过滤器，这为我们省去了很多麻烦。

````python
from filterpy.kalman import KalmanFilter
from filterpy.common import Q_discrete_white_noise

def pos_vel_filter(x, P, R, Q=0., dt=1.0):
    """ Returns a KalmanFilter which implements a
    constant velocity model for a state [x dx].T
    """
    
    kf = KalmanFilter(dim_x=2, dim_z=1)
    kf.x = np.array([x[0], x[1]]) # location and velocity
    kf.F = np.array([[1., dt],
                     [0.,  1.]])  # state transition matrix
    kf.H = np.array([[1., 0]])    # Measurement function
    kf.R *= R                     # measurement uncertainty
    if np.isscalar(P):
        kf.P *= P                 # covariance matrix 
    else:
        kf.P[:] = P               # [:] makes deep copy
    if np.isscalar(Q):
        kf.Q = Q_discrete_white_noise(dim=2, dt=dt, var=Q)
    else:
        kf.Q[:] = Q
    return kf
````

`KalmanFilter` 将 `R`、`P` 和 `Q` 初始化为单位矩阵，因此 `kf.P *= P` 是一种快速将所有对角线元素分配给相同标量值的方法。 现在我们创建过滤器：

````python
dt = .1
x = np.array([0., 0.]) 
kf = pos_vel_filter(x, P=500, R=5, Q=0.1, dt=dt)
````

````python
from kf_book.mkf_internal import plot_track

def run(x0=(0.,0.), P=500, R=0, Q=0, dt=1.0, 
        track=None, zs=None,
        count=0, do_plot=True, **kwargs):
    """
    track is the actual position of the dog, zs are the 
    corresponding measurements. 
    """

    # Simulate dog if no data provided. 
    if zs is None:
        track, zs = compute_dog_data(R, Q, count)

    # create the Kalman filter
    kf = pos_vel_filter(x0, R=R, P=P, Q=Q, dt=dt)  

    # run the kalman filter and store the results
    xs, cov = [], []
    for z in zs:
        kf.predict()
        kf.update(z)
        xs.append(kf.x)
        cov.append(kf.P)

    xs, cov = np.array(xs), np.array(cov)
    if do_plot:
        plot_track(xs[:, 0], track, zs, cov, **kwargs)
    return xs, cov
````

```python
    xs, cov = np.array(xs), np.array(cov)
```

### Prediction Equations

卡尔曼滤波器使用这些方程来计算 *prior* - 系统的预测下一个状态。 他们计算系统的先验均值 ($\bar{\mathbf x}$) 和协方差 ($\bar{\mathbf P}$)。

$$\begin{aligned}
\mathbf{\bar x} &= \mathbf{Fx} + \mathbf{Bu}\\
\mathbf{\bar P} &= \mathbf{FPF}^\mathsf T + \mathbf Q
\end{aligned}$$

$\underline{\textbf{Mean}}$

$\mathbf{\bar x} = \mathbf{Fx} + \mathbf{Bu}$

提醒一下，线性方程 $\mathbf{Ax} = \mathbf b$ 表示方程组，其中 $\mathbf A$ 保存方程组的系数，$\mathbf x$ 是变量向量。 执行乘法 $\mathbf{Ax}$ 计算该组方程的右侧值，由 $\mathbf b$ 表示。

如果 $\mathbf F$ 包含给定时间步的状态转换，则乘积 $\mathbf{Fx}$ 计算该转换后的状态。 简单！ 同样，$\mathbf B$ 是控制函数，$\mathbf u$ 是控制输入，所以 $\mathbf{Bu}$ 计算了转换后控制对状态的贡献。 因此，先验 $\mathbf{\bar x}$ 被计算为 $\mathbf{Fx}$ 和 $\mathbf{Bu}$ 之和。

等效的单变量方程是

$$\bar\mu = \mu + \mu_{move}$$

如果你执行矩阵乘法 $\mathbf{Fx}$，它会为 $x$ 生成这个方程。

让我们明确这一点。 回忆上一章中 $\mathbf F$ 的值：

$$\mathbf F = \begin{bmatrix}1&\Delta t  \\ 0&1\end{bmatrix}$$

因此 $\mathbf{\bar x} = \mathbf{Fx}$ 对应于线性方程组：

$$\begin{cases}
\begin{aligned}
\bar x &= 1x + &\Delta t\, \dot x \\
\bar{\dot x} &=0x + &1\, \dot x
\end{aligned}
\end{cases}$$



$\underline{\textbf{Covariance}}$

$\mathbf{\bar P} = \mathbf{FPF}^\mathsf T + \mathbf Q$

这个等式不太容易理解，所以我们会花更多的时间在上面。

在这个方程的单变量版本中是：

$$\bar\sigma^2 = \sigma^2 + \sigma^2_{move}$$

我们将运动的方差添加到我们估计的方差中，以反映知识的损失。 我们需要在这里做同样的事情，除了多元高斯不是那么容易。

我们不能简单地写成 $\mathbf{\bar P} = \mathbf P + \mathbf Q$。 在多元高斯中，状态变量是*相关的*。 这意味着什么？ 我们对速度的了解并不完善，但我们将其添加到位置

$$\bar x = \dot x\Delta t + x$$

由于我们不完全了解 $\dot x$ 的值，因此总和 $\bar x = \dot x\Delta t + x$ 获得了不确定性。 因为位置和速度是相关的，我们不能简单地添加协方差矩阵。 例如，如果 $\mathbf P$ 和 $\mathbf Q$ 是对角矩阵，则总和也是对角矩阵。 但是我们知道位置与速度相关，因此非对角线元素应该是非零的。

正确的方程是

$$\mathbf{\bar P} = \mathbf{FPF}^\mathsf T + \mathbf Q$$

$\mathbf{ABA}^\mathsf T$ 形式的表达式在线性代数中很常见。 您可以将其视为通过外部术语*投影*中间术语。 我们将在本书的其余部分多次使用它。 我承认这对你来说可能是一个“神奇”的方程式。 让我们来探索一下。

当我们初始化 $\mathbf P$ 时

$$\mathbf P = \begin{bmatrix}\sigma^2_x & 0 \\ 0 & \sigma^2_v\end{bmatrix}$$

$\mathbf{FPF}^\mathsf T$ 的值为：

$$\begin{aligned}
\mathbf{FPF}^\mathsf T &= \begin{bmatrix}1&\Delta t\\0&1\end{bmatrix}
\begin{bmatrix}\sigma^2_x & 0 \\  0 & \sigma^2_{v}\end{bmatrix}
\begin{bmatrix}1&0\\\Delta t&1\end{bmatrix} \\
&= \begin{bmatrix}\sigma^2_x&\sigma_v^2\Delta t\\  0 & \sigma^2_{v}\end{bmatrix}
\begin{bmatrix}1&0\\\Delta t&1\end{bmatrix} \\
&= \begin{bmatrix}\sigma^2_x +  \sigma_v^2\Delta t^2  &  \sigma_v^2\Delta t \\
\sigma_v^2\Delta t & \sigma^2_{v}\end{bmatrix}
\end{aligned}$$

$\mathbf P$ 的初始值在位置和速度之间没有协方差。位置计算为 $\dot x\Delta t + x$，因此位置和速度之间存在相关性。乘法 $\mathbf{FPF}^\mathsf T$ 计算出 $\sigma_v^2 \Delta t$ 的协方差。确切的值并不重要；您只需要认识到 $\mathbf{FPF}^\mathsf T$ 使用过程模型来自动计算位置和速度之间的协方差！

另一种思考方式是考虑 $\mathbf{Fx}$ 乘法。那预计 $\mathbf x$ 及时向前。 $\mathbf {FP}$ 可能看起来是等价的运算，但 $\mathbf P$ 是一个矩阵，而 $\mathbf x$ 是一个向量。尾随的 $\mathbf F^\mathsf T$ 项确保我们乘以 $\mathbf F$ 的行和列。在 $\mathbf{FPF}^\mathsf T$ 计算的第二行中，我们得到了 $\mathbf{FP}$ 的值。您可以看到它是一个上三角矩阵，因为我们还没有将 $\mathbf F$ 完全合并到乘法中。

如果您有线性代数和统计方面的经验，这可能会有所帮助。由于预测的协方差可以建模为预测步骤中误差的预期值，由该方程给出。

$$\begin{aligned}
\bar{\mathbf P} &= \mathbb E[(\mathbf{Fx - \bar \mu})(\mathbf{Fx - \bar\mu})^\mathsf T]\\
 &= \mathbf F\, \mathbb E[\mathbf{(x- \bar\mu)(x- \bar\mu)}^\mathsf T]\, \mathbf F^\mathsf T
\end{aligned}$$

当然，$\mathbb E[\mathbf{(x- \bar\mu)(x- \bar\mu)}^\mathsf T]$ 只是 $\mathbf P$，给我们

$$\bar{\mathbf P} = \mathbf{FPF}^\mathsf T$$

我们来看看它的效果。 在这里，我使用过滤器中的 $\mathbf F$ 并将状态向前投射 6/10 秒。 我这样做了五次，这样你就可以看到 $\mathbf{\bar P}$ 是如何继续变化的。

````python
dt = 0.6
x = np.array([0., 5.])
F = np.array([[1., dt], [0, 1.]])
P = np.array([[1.5, 0], [0, 3.]])
plot_covariance_ellipse(x, P, edgecolor='r')

for _ in range(5):
    x = F @ x
    P = F @ P @ F.T
    plot_covariance_ellipse(x, P, edgecolor='k', ls='dashed')
book_plots.set_labels(x='position', y='velocity')
````

您可以看到，速度为 5 时，位置在每 6/10 秒内正确移动 3 个单位。在每一步，椭圆的宽度都变大了，这表明我们在每一步都向 x 添加了 $\dot x\Delta t$，从而丢失了关于位置的信息。高度没有改变——我们的系统模型说速度没有改变，所以我们对速度的信念不会改变。随着时间的推移，您可以看到椭圆变得越来越倾斜。回想一下，倾斜表示*相关*。 $\mathbf F$ 将 $x$ 与 $\dot x$ 与表达式 $\bar x = \dot x \Delta t + x$ 线性相关。 $\mathbf{FPF}^\mathsf T$ 计算正确地将这种相关性合并到协方差矩阵中。

### Update Equations

更新方程看起来比预测方程更混乱，但这主要是由于卡尔曼滤波器计算了测量空间中的更新。 这是因为测量值不是*可逆的*。 例如，考虑一个为目标提供范围的传感器。 将范围转换为位置是不可能的 - 一个圆圈中的无限数量的位置将产生相同的范围。 另一方面，我们总是可以计算给定位置（状态）的范围（测量值）。

在我继续之前，回想一下我们正在尝试做一些非常简单的事情：选择一个介于测量和预测之间的新估计，如下图所示：

<img src="./figs/residual_chart.png">

方程会很复杂，因为状态有多个维度，但我们正在做这个操作。 不要让方程式分散您对这个想法的简单性的注意力。

$\underline{\textbf{System Uncertainty}}$

$\textbf{S} = \mathbf{H\bar PH}^\mathsf T + \mathbf R$

为了在测量空间中工作，卡尔曼滤波器必须将协方差矩阵投影到测量空间中。 其数学公式为 $\mathbf{H\bar PH}^\mathsf T$，其中 $\mathbf{\bar P}$ 是*先验*协方差，$\mathbf H$ 是测量函数。


你应该认出这个 $\mathbf{ABA}^\mathsf T$ 形式 - 预测步骤使用 $\mathbf{FPF}^\mathsf T$ 用状态转换函数更新 $\mathbf P$。 在这里，我们使用相同的表格来使用测量功能对其进行更新。 线性代数正在为我们改变坐标系。

一旦协方差在测量空间中，我们需要考虑传感器噪声。 这很容易——我们只需添加矩阵。 结果被称为*系统不确定性*或*创新协方差*。

如果你忽略 $\mathbf H$ 项，这个方程等价于卡尔曼增益的单变量方程中的分母：


$$K = \frac {\bar\sigma^2} {\bar\sigma^2 + \sigma_z^2}$$

比较系统不确定性和协方差的方程

$$\begin{aligned}
\mathbf{S} &= \mathbf{H\bar PH}^\mathsf T + \mathbf R\\
\mathbf{\bar P} &= \mathbf{FPF}^\mathsf T + \mathbf Q
\end{aligned}$$

在每个方程中，$\mathbf P$ 用函数 $\mathbf H$ 或 $\mathbf F$ 放入不同的空间。 然后我们添加与该空间相关的噪声矩阵。

$\underline{\textbf{Kalman Gain}}$

$\mathbf K = \mathbf{\bar PH}^\mathsf T \mathbf{S}^{-1}$

回头看残差图。 一旦我们有了预测和测量，我们就需要在两者之间选择一个估计值。 如果我们对测量有更多的确定性，估计会更接近它。 相反，如果我们对预测有更多的确定性，那么估计将更接近它。

在单变量章节中，我们使用这个等式缩放了平均值

$$
\mu =\frac{\bar\sigma^2 \mu_z + \sigma_\mathtt{z}^2 \bar\mu} {\bar\sigma^2 + \sigma_\mathtt{z}^2}$$

which we simplified to

$$\mu = (1-K)\bar\mu + K\mu_\mathtt{z}$$

这给了我们

$$K = \frac {\bar\sigma^2} {\bar\sigma^2 + \sigma_z^2}$$

$K$ 是*卡尔曼增益*，它是介于 0 和 1 之间的实数。确保您了解它如何在预测和测量之间选择平均值。 卡尔曼增益是 *percentage* 或 *ratio* - 如果 K 为 0.9，则需要 90% 的测量值和 10% 的预测值。

对于多元卡尔曼滤波器 $\mathbf K$ 是一个向量，而不是一个标量。 这里还是这个等式：$\mathbf K = \mathbf{\bar PH}^\mathsf T \mathbf{S}^{-1}$。 这是*比率*吗？ 我们可以将矩阵的逆视为线性代数求倒数的方式。 没有为矩阵定义除法，但以这种方式考虑它是有用的。 所以我们可以将 $\textbf{K}$ 的等式解读为含义

$$\begin{aligned} \mathbf K &\approx \frac{\mathbf{\bar P}\mathbf H^\mathsf T}{\mathbf{S}} \\
\mathbf K &\approx \frac{\mathsf{uncertainty}_\mathsf{prediction}}{\mathsf{uncertainty}_\mathsf{prediction} + \mathsf{uncertainty}_\mathsf{measurement}}\mathbf H^\mathsf T
\end{aligned}$$


卡尔曼增益方程根据我们对预测与测量的信任程度来计算比率。 我们在前面的每一章都做了同样的事情。 这个方程很复杂，因为我们是通过矩阵在多个维度上做的，但这个概念很简单。 $\mathbf H^\mathsf T$ 术语不太清楚，我会尽快解释。 如果忽略该项，卡尔曼增益的方程与单变量情况相同：将先验的不确定性除以先验和测量的不确定性之和。

$\underline{\textbf{Residual}}$

$\mathbf y = \mathbf z - \mathbf{H\bar{x}}$


这是一个简单的问题，因为我们在设计测量函数 $\mathbf H$ 时已经涵盖了这个等式。 回想一下，测量函数将状态转换为测量值。 因此 $\mathbf{Hx}$ 将 $\mathbf x$ 转换为等效的测量值。 完成后，我们可以从测量值 $\mathbf z$ 中减去它以获得残差 - 测量值和预测值之间的差异。

单变量方程是


$$y = z - \bar x$$

并且清楚地计算了同样的事情，但只是在一维上。

$\underline{\textbf{State Update}}$

$\mathbf x = \mathbf{\bar x} + \mathbf{Ky}$

我们选择我们的新状态沿着残差，由卡尔曼增益缩放。 缩放由 $\mathbf{Ky}$ 执行，它缩放残差并将其转换回状态空间，其中 $\mathbf H^\mathsf T$ 项位于 $\mathbf K$ 中。 这被添加到先验中，得到等式：$\mathbf x =\mathbf{\bar x} + \mathbf{Ky}$。 让我写出 $\mathbf K$ 以便我们可以看到整个计算：

$$\begin{aligned}
\mathbf x &= \mathbf{\bar x} + \mathbf{Ky} \\
&= \mathbf{\bar x} + \mathbf{\bar PH}^\mathsf T \mathbf{S}^{-1}\mathbf y \\
&\approx \mathbf{\bar x} + \frac{\mathsf{uncertainty}_\mathsf{prediction}}{\mathsf{uncertainty}_\mathsf{measurement}}\mathbf H^\mathsf T\mathbf y
\end{aligned}$$

也许*看到*比率的更好方法是重写估计方程：


$$\begin{aligned}
\mathbf x &= \mathbf{\bar x} + \mathbf{Ky} \\
&= \mathbf{\bar x} +\mathbf K(\mathbf z - \mathbf{H\bar x}) \\
&= (\mathbf I - \mathbf{KH})\mathbf{\bar x} + \mathbf{Kz}
\end{aligned}$$

这与单变量形式之间的相似性应该是显而易见的：

$$\mu = (1-K)\bar\mu + K\mu_\mathtt{z}$$

$\underline{\textbf{Covariance Update}}$

$\mathbf P = (\mathbf{I}-\mathbf{KH})\mathbf{\bar P}$


$\mathbf{I}$ 是单位矩阵，也是我们在多个维度上表示 $1$ 的方式。 $\mathbf H$ 是我们的测量函数，是一个常数。 我们可以将等式视为 $\mathbf P = (1-c\mathbf K)\mathbf P$。 $\mathbf K$ 是我们使用多少预测与测量的比率。 如果 $\mathbf K$ 很大，则 $(1-\mathbf{cK})$ 很小，并且 $\mathbf P$ 将比原来更小。 如果$\mathbf K$ 较小，则$(1-\mathbf{cK})$ 较大，$\mathbf P$ 会相对较大。 这意味着我们通过卡尔曼增益的某些因素来调整不确定性的大小。

这个方程在数值上可能不稳定，我不在 FilterPy 中使用它。 随着时间的推移，减法会破坏对称性并导致浮点错误。 稍后我将分享这个方程的更复杂但数值稳定的形式。

### An Example not using FilterPy

FilterPy 对我们隐藏了实现的细节。 通常你会喜欢这一点，但让我们在没有 FilterPy 的情况下实现最后一个过滤器。 为此，我们需要将矩阵定义为变量，然后显式实现卡尔曼滤波器方程。

在这里我们初始化我们的矩阵：

````python
dt = 1.
R_var = 10
Q_var = 0.01
x = np.array([[10.0, 4.5]]).T
P = np.diag([500, 49])
F = np.array([[1, dt],
              [0,  1]])
H = np.array([[1., 0.]])
R = np.array([[R_var]])
Q = Q_discrete_white_noise(dim=2, dt=dt, var=Q_var)
````

````python
from scipy.linalg import inv

count = 50
track, zs = compute_dog_data(R_var, Q_var, count)
xs, cov = [], []
for z in zs:
    # predict
    x = F @ x
    P = F @ P @ F.T + Q
    
    #update
    S = H @ P @ H.T + R
    K = P @ H.T @ inv(S)
    y = z - H @ x
    x += K @ y
    P = P - K @ H @ P
    
    xs.append(x)
    cov.append(P)

xs, cov = np.array(xs), np.array(cov)
plot_track(xs[:, 0], track, zs, cov, plot_P=False)
````

更重要的是，此示例要求您记住卡尔曼滤波器的方程并对其进行编程。 迟早你会犯错。 FilterPy 的版本确保您的代码是正确的。 另一方面，如果您在定义中犯了错误，例如将 $\mathbf H$ 设为列向量而不是行向量，则 FilterPy 的错误消息将比此显式代码更难调试。

FilterPy 的 KalmanFilter 类提供了额外的功能，例如平滑、批处理、褪色记忆过滤、最大似然函数的计算等。 您无需显式编程即可获得所有这些功能。

$$
\begin{aligned}
\text{Predict Step}\\
\mathbf{\bar x} &= \mathbf{F x} + \mathbf{B u} \\
\mathbf{\bar P} &= \mathbf{FP{F}}^\mathsf T + \mathbf Q \\
\\
\text{Update Step}\\
\textbf{S} &= \mathbf{H\bar PH}^\mathsf T + \mathbf R \\
\mathbf K &= \mathbf{\bar PH}^\mathsf T \mathbf{S}^{-1} \\
\textbf{y} &= \mathbf z - \mathbf{H \bar x} \\
\mathbf x &=\mathbf{\bar x} +\mathbf{K\textbf{y}} \\
\mathbf P &= (\mathbf{I}-\mathbf{KH})\mathbf{\bar P}
\end{aligned}
$$

我想分享一种你将在文献中看到的方程形式。 使用了许多不同的符号系统，但这可以让您了解预期的内容。

 $$
\begin{aligned}
\hat{\mathbf x}_{k\mid k-1} &= \mathbf F_k\hat{\mathbf x}_{k-1\mid k-1} + \mathbf B_k \mathbf u_k  \\
\mathbf P_{k\mid k-1} &=  \mathbf F_k \mathbf P_{k-1\mid k-1} \mathbf F_k^\mathsf T + \mathbf Q_k \\        	
\tilde{\mathbf y}_k &= \mathbf z_k - \mathbf H_k\hat{\mathbf x}_{k\mid k-1}\\
\mathbf{S}_k &= \mathbf H_k \mathbf P_{k\mid k-1} \mathbf H_k^\mathsf T + \mathbf R_k \\
\mathbf K_k &= \mathbf P_{k\mid k-1}\mathbf H_k^\mathsf T \mathbf{S}_k^{-1}\\
\hat{\mathbf x}_{k\mid k} &= \hat{\mathbf x}_{k\mid k-1} + \mathbf K_k\tilde{\mathbf y}_k\\
\mathbf P_{k|k} &= (I - \mathbf K_k \mathbf H_k) \mathbf P_{k|k-1}
\\\end{aligned}
$$

这个符号使用贝叶斯 $a\mid b$ 符号，这意味着 $a$ 给定了 $b$ 的证据。帽子意味着估计。因此，$\hat{\mathbf x}_{k\mid k}$ 表示在给定步骤 $k$（第二个 k ）。换句话说，后面。 $\hat{\mathbf x}_{k\mid k-1}$ 表示状态 $\mathbf x$ 在步骤 $k$ 的估计，给定步骤 $k - 1$ 的估计。换句话说，先验。

编写一个使用状态 $\mathbf x=\begin{bmatrix}x\end{bmatrix}$ 的卡尔曼滤波器，并将其与使用 $\mathbf x=\begin{bmatrix} x & \dot x \end 的滤波器进行比较 {bmatrix}^\mathsf T $。

### Solution

我们已经为位置和速度实现了卡尔曼滤波器，所以我将提供代码而不做太多评论，然后绘制结果。

````python
from math import sqrt
from numpy.random import randn

def univariate_filter(x0, P, R, Q):
    f = KalmanFilter(dim_x=1, dim_z=1, dim_u=1)
    f.x = np.array([[x0]])
    f.P *= P
    f.H = np.array([[1.]])
    f.F = np.array([[1.]])
    f.B = np.array([[1.]])
    f.Q *= Q
    f.R *= R
    return f

def plot_1d_2d(xs, xs1d, xs2d):
    plt.plot(xs1d, label='1D Filter')
    plt.scatter(range(len(xs2d)), xs2d, c='r', alpha=0.7, label='2D Filter')
    plt.plot(xs, ls='--', color='k', lw=1, label='track')
    plt.title('State')
    plt.legend(loc=4)
    plt.show()
    
def compare_1D_2D(x0, P, R, Q, vel, u=None):
    # storage for filter output
    xs, xs1, xs2 = [], [], []

    # 1d KalmanFilter
    f1D = univariate_filter(x0, P, R, Q)

    #2D Kalman filter
    f2D = pos_vel_filter(x=(x0, vel), P=P, R=R, Q=0)
    if np.isscalar(u):
        u = [u]
    pos = 0 # true position
    for i in range(100):
        pos += vel
        xs.append(pos)

        # control input u - discussed below
        f1D.predict(u=u)
        f2D.predict()
        
        z = pos + randn()*sqrt(R) # measurement
        f1D.update(z)
        f2D.update(z)
        
        xs1.append(f1D.x[0])
        xs2.append(f2D.x[0])
    plt.figure()
    plot_1d_2d(xs, xs1, xs2)

compare_1D_2D(x0=0, P=50., R=5., Q=.02, vel=1.) 
````

## How Velocity is Calculated

我还没有解释过滤器如何计算速度或任何隐藏变量。 如果我们插入我们为每个过滤器矩阵计算的值，我们可以看到会发生什么。

首先我们需要计算系统的不确定性。

$$\begin{aligned}
\textbf{S} &= \mathbf{H\bar PH}^\mathsf T + \mathbf R \\
&= \begin{bmatrix} 1 & 0\end{bmatrix}
\begin{bmatrix}\sigma^2_x & \sigma_{xv} \\ \sigma_{xv} & \sigma^2_v\end{bmatrix}
\begin{bmatrix} 1 \\ 0\end{bmatrix} + \begin{bmatrix}\sigma_z^2\end{bmatrix}\\
&= \begin{bmatrix}\sigma_x^2 & \sigma_{xv}\end{bmatrix}\begin{bmatrix} 1 \\ 0\end{bmatrix}+ \begin{bmatrix}\sigma_z^2\end{bmatrix} \\
&= \begin{bmatrix}\sigma_x^2 +\sigma_z^2\end{bmatrix}
\end{aligned}$$

现在我们有了 $\mathbf S$，我们可以找到卡尔曼增益的值：

$$\begin{aligned}
\mathbf K &= \mathbf{\bar PH}^\mathsf T \mathbf{S}^{-1} \\
&= \begin{bmatrix}\sigma^2_x & \sigma_{xv} \\ \sigma_{xv} & \sigma^2_v\end{bmatrix}
\begin{bmatrix} 1 \\ 0\end{bmatrix}
\begin{bmatrix}\frac{1}{\sigma_x^2 +\sigma_z^2}\end{bmatrix} \\
&= \begin{bmatrix}\sigma^2_x  \\ \sigma_{xv}\end{bmatrix}
\begin{bmatrix}\frac{1}{\sigma_x^2 +\sigma_z^2}\end{bmatrix} \\
&= \begin{bmatrix}\sigma^2_x/(\sigma_x^2 +\sigma_z^2)  \\ \sigma_{xv}/(\sigma_x^2 +\sigma_z^2)\end{bmatrix}
\end{aligned}
$$

换句话说，$x$ 的卡尔曼增益为

$$K_x = \frac{VAR(x)}{VAR(x)+VAR(z)}$$

从单变量的情况来看，这对您来说应该非常熟悉。

速度 $\dot x$ 的卡尔曼增益为
$$K_{\dot x} = \frac{COV(x, \dot x)}{VAR(x)+VAR(z)}$$

这有什么作用？ 回想一下，我们将状态计算为

$$\begin{aligned}\mathbf x 
&=\mathbf{\bar x}+\mathbf K(z-\mathbf{Hx)} \\
&= \mathbf{\bar x}+\mathbf Ky\end{aligned}$$

这里的残差 $y$ 是一个标量。 因此它被乘以$\mathbf K$ 的每个元素。 因此我们有

$$\begin{bmatrix}x \\ \dot x\end{bmatrix}=\begin{bmatrix}\bar x \\ \bar{\dot x}\end{bmatrix} + \begin{bmatrix}K_x \\ K_{\dot x}\end{bmatrix}y$$

这给出了这个方程组：

$$\begin{aligned}x& = \bar x + yK_x\\
\dot x &= \bar{\dot x} + yK_{\dot x}\end{aligned}$$

预测 $\bar x$ 计算为 $x + \bar x \Delta t$。如果预测是完美的，那么残差将是 $y=0$（忽略测量中的噪声）并且速度估计将保持不变。另一方面，如果速度估计很差，那么预测会很差，残差会很大：$y >> 0$。在这种情况下，我们用 $yK_{\dot x}$ 更新速度估计。 $K_{\dot x}$ 与 $COV(x,\dot x)$ 成正比。因此，速度通过位置误差乘以与位置和速度之间的协方差成比例的值来更新。相关性越高，修正越大。

为了得到这个完整的循环，$COV(x,\dot x)$ 是 $\mathbf P$ 的非对角元素。回想一下，这些值是用 $\mathbf{FPF}^\mathsf T$ 计算的。因此，位置和速度的协方差是在预测步骤中计算的。速度的卡尔曼增益与该协方差成正比，我们根据在最后一个时期的不准确程度乘以与该协方差成正比的值来调整速度估计。

综上所述，这些线性代数方程你可能不熟悉，但计算其实很简单。它本质上与我们在 g-h 滤波器中执行的计算相同。本章中的常数不同，因为我们考虑了过程模型和传感器中的噪声，但数学是相同的。

查看输出，我们在开始时看到滤波器输出中有一个非常大的尖峰。 我们设置 $\text{P}=500\, \mathbf{I}_2$（这是对角线为 500 的 2x2 对角矩阵的简写符号）。 我们现在有足够的信息来理解这意味着什么，以及卡尔曼滤波器如何处理它。 左上角的500对应$\sigma^2_x$； 因此我们说 `x` 的标准差是 $\sqrt{500}$，或大约 22.36 m。 大约 99% 的样本出现在 $3\sigma$ 内，因此 $\sigma^2_x=500$ 告诉卡尔曼滤波器预测（先验）可能高达 67 米。 这是一个很大的误差，所以当测量值达到峰值时，卡尔曼滤波器不信任自己的估计值，并疯狂地跳跃以尝试合并测量值。 然后，随着过滤器的发展，$\mathbf P$ 迅速收敛到更现实的值。

让我们看看这背后的数学原理。 卡尔曼增益方程为

$$\mathbf K = \mathbf{\bar P} \mathbf H^\mathsf T\mathbf{S}^{-1} \approx \frac{\mathbf{\bar P}\mathbf H^\mathsf T}{\mathbf{S}} 
\approx \frac{\mathsf{uncertainty}_\mathsf{prediction}}{\mathsf{uncertainty}_\mathsf{measurement}}\mathbf H^\mathsf T
$$

它是预测与测量的不确定性之比。 这里预测的不确定性很大，所以 $\mathbf K$ 很大（如果这是一个标量，则接近 1）。 $\mathbf K$ 乘以残差 $\textbf{y} = \mathbf z - \mathbf{H \bar x}$，即测量值减去预测值，因此较大的 $\mathbf K$ 有利于测量值 . 因此，如果 $\mathbf P$ 相对于传感器不确定性 $\mathbf R$ 较大，则滤波器将形成来自测量的大部分估计。


现在让我们看看 $\mathbf P = 1.0\, \mathbf{I}_2$ 的较小初始值的效果。

````python
run(track=trk, zs=zs, R=var, Q=.02, P=1., plot_P=True, 
    title='$P=1\, m^2$');
````

这*看起来*乍一看还不错。该情节没有前一个情节的尖峰；滤波器开始跟踪测量结果，不需要任何时间来稳定信号。但是，如果我们查看 P 的图，您会发现位置方差存在初始峰值，并且它从未真正收敛。糟糕的设计会导致较长的收敛时间和次优结果。

因此，尽管滤波器跟踪非常接近实际信号，但我们不能断定“魔术”是使用一个小的 $\mathbf P$。是的，这将避免卡尔曼滤波器花费时间来准确跟踪信号，但如果我们对初始测量确实不确定，这可能会导致滤波器产生非常糟糕的结果。如果我们正在追踪一个活的物体，在我们开始追踪它之前，我们可能非常不确定它在哪里。另一方面，如果我们过滤温度计的输出，我们对第 1000 次测量结果的确定性和第一次测量结果一样确定。为了使您的卡尔曼滤波器表现良好，您必须将 $\mathbf P$ 设置为真正反映您对数据的了解的值。

## Question: Explain Ellipse Differences

为什么 $\mathbf R=5 m^2$ 的椭圆比 $\mathbf R=0.5 m^2$ 的椭圆更倾向于水平方向。 提示：考虑这些椭圆的物理含义，而不是数学。 如果您不确定答案，请将 $\mathbf R$ 更改为真正的大小数字，例如 100 $m^2$ 和 0.1 $m^2$，观察变化，并思考这意味着什么。

### Solution

x 轴代表位置，y 轴代表速度。垂直或几乎垂直的椭圆表示位置和速度之间没有相关性，而对角线的椭圆表示存在很多相关性。这样说来，结果听起来不太可能。椭圆的倾斜度会发生变化，但相关性不应随时间而变化。但这是对*过滤器输出*的度量，而不是对实际物理世界的描述。当 $\mathbf R$ 非常大时，我们告诉滤波器测量中有很多噪声。在这种情况下，卡尔曼增益 $\mathbf K$ 被设置为有利于预测而不是测量，并且预测来自速度状态变量。因此，$x$ 和 $\dot x$ 之间存在很大的相关性。相反，如果 $\mathbf R$ 很小，我们告诉过滤器测量值非常可靠，并且 $\mathbf K$ 设置为有利于测量而不是预测。如果测量接近完美，为什么过滤器要使用预测？如果过滤器没有从预测中使用太多，则报告的相关性将非常少。

**这是理解的关键点！**。卡尔曼滤波器是真实世界系统的数学模型。几乎没有相关性的报告*并不意味着*物理系统中没有相关性，只是在数学模型中没有*线性*相关性。它是关于模型中包含多少测量与预测的报告。

让我们用一个真正大的测量误差来说明这一点。我们将设置 $\mathbf R=200\, m^2$。在看之前想想情节会是什么样子。

````python
plot_covariances(R=200., Q=.2, count=5, title='$R = 200\, m^2$')
````

## Filter Initialization

有许多用于初始化过滤器的方案。 以下方法在大多数情况下表现良好。 在这个方案中，在获得第一个测量值 $\mathbf z_0$ 之前，您不会初始化滤波器。 由此你可以用 $\mathbf x_0 = \mathbf z_0$ 计算 $\mathbf x$ 的初始值。 如果 $\mathbf z$ 与 $\mathbf x$ 的大小、类型和单位不同（通常是这种情况），我们可以使用如下测量函数。

我们知道

$$\mathbf z = \mathbf{Hx}$$

因此

$$\begin{aligned}
\mathbf H^{-1}\mathbf{Hx} &= \mathbf H^{-1}\mathbf z \\
\mathbf x &= \mathbf H^{-1}\mathbf z\end{aligned}$$

矩阵求逆需要方阵，但 $\mathbf H$ 很少是方阵。 SciPy 将使用 scipy.linalg.pinv 计算矩阵的 Moore-Penrose 伪逆，因此您的代码可能看起来像

````python
from scipy.linalg import pinv

H = np.array([[1, 0.]]) 
z0 = 3.2
x = np.dot(pinv(H), z0)
print(x)
````

对您的问题领域的专业知识可能会导致您进行不同的计算，但这是一种方法。 例如，如果状态包括速度，您可以进行前两次位置测量，计算差值，并将其用作初始速度。

现在我们需要计算 $\mathbf P$ 的值。 这会因问题而异，但通常您将使用测量误差 $\mathbf R$ 用于相同的项，并为其余项使用最大值。 也许这还不清楚。 在本章中，我们一直在使用位置和速度作为状态进行跟踪和对象，并且测量结果是位置。 在这种情况下，我们将初始化 $\mathbf P$

$$\mathbf P = \begin{bmatrix}\mathbf R_0 & 0 \\0 & vel_{max}^2\end{bmatrix}$$

$\mathbf P$ 的对角线包含每个状态变量的方差，因此我们用合理的值填充它。 $\mathbf R$ 是位置的合理方差，最大速度平方是速度的合理方差。 它是平方的，因为方差是平方的：$\sigma^2$。

您确实需要了解您正在工作的领域，并根据最佳可用信息初始化您的过滤器。 例如，假设我们试图在赛马中追踪马匹。 初始测量可能非常糟糕，并为您提供远离起始门的位置。 我们知道马必须从起跑门出发； 将滤波器初始化为初始测量值会导致次优结果。 在这种情况下，我们希望始终使用马的起始门位置初始化卡尔曼滤波器。

## Batch Processing

卡尔曼滤波器被设计为一种递归算法——当新的测量值出现时，我们立即创建一个新的估计值。 但是，我们想要过滤一组已经收集的数据是很常见的。 卡尔曼滤波器可以在批处理模式下运行，其中所有测量值都被一次过滤。 我们已经在 `KalmanFilter.batch_filter()` 中实现了这一点。 在内部，该函数所做的只是循环测量并在数组中收集结果状态和协方差估计。 它简化了您的逻辑并方便地将所有输出收集到数组中。 我经常使用这个函数，但一直等到本章结束，这样你就会非常熟悉必须运行的预测/更新周期。

首先将您的测量值收集到一个数组或列表中。 也许它在一个 CSV 文件中：

```python
zs = read_altitude_from_csv('altitude_data.csv')
```

或者，也许您将使用生成器生成它：

```python
zs = [some_func(i) for i in range(1000)]
```

然后调用 `batch_filter()` 方法。

```python
Xs, Ps, Xs_prior, Ps_prior = kfilter.batch_filter(zs)
```

该函数获取测量值列表，对其进行过滤，并返回状态估计 (Xs)、协方差矩阵 (Ps) 和相同的先验 (Xs_prior, Ps_prior) 的 NumPy 数组。

这是一个完整的例子。

````python
count = 50
track, zs = compute_dog_data(10, .2, count)
P = np.diag([500., 49.])
f = pos_vel_filter(x=(0., 0.), R=3., Q=.02, P=P)
xs, _, _, _ = f.batch_filter(zs)

book_plots.plot_measurements(range(1, count + 1), zs)
book_plots.plot_filter(range(1, count + 1), xs[:, 0])
plt.legend(loc='best');
````

批处理过滤器采用可选的 `filterpy.common.Saver` 对象。 如果提供，过滤器的所有属性也将被保存。 如果您想检查状态和协方差以外的值，这很有用。 在这里，我绘制了残差，看看它是否看起来像以 0 为中心的噪声。这是一个快速的视觉检查，看看滤波器是否设计得很好。 如果如果从零漂移，或者看起来不像噪声，则滤波器设计不良和/或过程不是高斯的。 我们将在后面的章节中详细讨论这一点。 现在考虑这是一个 `Saver` 类的演示。

````python
track, zs = compute_dog_data(10, .2, 200)
P = np.diag([500., 49.])
f = pos_vel_filter(x=(0., 0.), R=3., Q=.02, P=P)
s = Saver(f)
xs, _, _, _ = f.batch_filter(zs, saver=s)
s.to_array()
plt.plot(s.y);
````

## Smoothing the Results

这本书包括一章关于平滑结果的章节；我不会在这里重复信息。但是，它非常易于使用，并且提供了如此深刻的改进输出，我将用几个例子来逗你。平滑章节不是特别难；你现在已经准备好阅读它了。

假设我们正在跟踪一辆直线行驶的汽车。我们得到的测量结果表明汽车开始向左转。卡尔曼滤波器将状态估计稍微移向测量，但它无法判断这是一个特别嘈杂的测量还是一个转弯的真正开始。

但是，如果我们有未来的测量结果，我们可以决定是否进行了转弯。假设随后的测量都继续左转。然后我们可以确定一个转弯已经开始。另一方面，如果随后的测量继续沿直线进行，我们就会知道测量是有噪声的，应该大部分被忽略。不是在测量和预测之间进行估计，而是完全结合测量或忽略它，这取决于未来测量对物体运动的暗示。

`KalmanFilter` 实现了这种算法的一种形式，称为 *RTS 平滑器*，以算法的发明者 Rauch、Tung 和 Striebel 的名字命名。该方法是`rts_smoother()`。要使用它，传入从“batch_filter”步骤计算的均值和协方差，并接收平滑后的均值、协方差和卡尔曼增益。

````python
from numpy.random import seed
count = 50
seed(8923)

P = np.diag([500., 49.])
f = pos_vel_filter(x=(0., 0.), R=3., Q=.02, P=P)
track, zs = compute_dog_data(3., .02, count)
Xs, Covs, _, _ = f.batch_filter(zs)
Ms, Ps, _, _ = f.rts_smoother(Xs, Covs)

book_plots.plot_measurements(zs)
plt.plot(Xs[:, 0], ls='--', label='Kalman Position')
plt.plot(Ms[:, 0], label='RTS Position')
plt.legend(loc=4);
````

这个输出太棒了！ 在这张图表中，有两件事对我来说非常明显。 首先，RTS 平滑器的输出比 KF 输出平滑得多。 其次，它几乎总是比 KF 输出更准确（我们将在 **Smoothing** 章节中详细检查此声明）。 作为隐藏变量的速度的提升更为显着：

````python
plt.plot(Xs[:, 1], ls='--', label='Kalman Velocity')
plt.plot(Ms[:, 1], label='RTS Velocity')
plt.legend(loc=4)
plt.gca().axhline(1, lw=1, c='k');
````

## Discussion and Summary

多元高斯允许我们同时处理多个维度，包括空间维度和其他维度（速度等）。我们得出了一个关键见解：隐藏变量能够显着提高过滤器的准确性。这是可能的，因为隐藏变量与观察变量相关。

我给出了*可观察性*的直观定义。可观测性是由 Kalman 博士为线性系统发明的，其背后有相当多的理论。它回答了是否可以通过观察系统的输出来确定系统状态的问题。对于我们的问题，这很容易确定，但更复杂的系统可能需要严格的分析。维基百科的【可观察性】（https://en.wikipedia.org/wiki/Observability）文章有概述；如果您需要学习主题 [[Grewal2008]](#References) 是一个很好的来源。

关于隐藏变量有一个重要的警告。很容易构建一个过滤器来产生隐藏变量的估计。我可以编写一个过滤器来估计履带车的颜色。但是没有办法从位置计算汽车颜色，所以颜色的估计是无稽之谈。设计者必须验证这些变量的估计是否正确。如果您没有速度传感器但正在估算速度，则需要测试速度估算值是否正确。不要相信他们是。例如，假设速度有一个周期性分量——它看起来像一个正弦波。如果您的采样时间小于频率的 2 倍，您将无法准确估计速度（由于 Nyquist 定理）。假设采样周期等于速度的频率。滤波器将报告速度是恒定的，因为它在正弦波的同一点对系统进行采样。

初始化对隐藏变量提出了一个特别困难的问题。如果您从错误的初始化开始，过滤器通常可以恢复观察到的变量，但可能会因隐藏变量而挣扎并失败。估计隐藏变量是一种强大的工具，但也是一种危险的工具。

我建立了一系列设计卡尔曼滤波器的步骤。这些不是卡尔曼滤波器文献的常见部分，仅作为指导而非处方。针对难题进行设计是一个迭代过程。您对状态向量进行猜测，计算出您的测量和状态模型是什么，运行一些测试，然后根据需要更改设计。

$\mathbf R$ 和 $\mathbf Q$ 的设计通常非常具有挑战性。我已经让它看起来很科学。你的传感器有 $\mathcal{N}(0, \sigma^2)$ 的高斯噪声，所以设置 $\mathbf R=\sigma^2$。简单！这是一个肮脏的谎言。传感器不是高斯的。我们从一个浴室秤开始这本书。假设 $\sigma=1$ kg，并且您尝试称重 0.5 kg 的东西。理论告诉我们，我们会得到负测量值，但当然秤永远不会报告小于零的重量。现实世界的传感器通常具有 *fat tails*（称为 *kurtosis*）和 *skew*。在某些情况下，例如使用比例尺，一个或两个尾部会被截断。

$\mathbf Q$ 的情况更可怕。当我愉快地为我对狗的运动的预测分配一个噪声矩阵时，我希望你持怀疑态度。谁能说狗接下来会做什么？我的 GPS 中的卡尔曼滤波器不了解山丘、外部风或我糟糕的驾驶技术。然而，过滤器需要一个精确的数字来封装所有这些信息，并且当我在沙漠中越野驾驶时，以及当一级方程式冠军在赛道上驾驶时，它需要工作。

这些问题导致一些研究人员和工程师贬低地将卡尔曼滤波器称为“泥球”。换句话说，它并不总是能很好地结合在一起。要知道的另一个术语 - 卡尔曼滤波器可以变成 *smug*。他们的估计完全基于你告诉它的噪音。这些值可能导致过度自信的估计。 $\mathbf P$ 变得越来越小，而过滤器实际上变得越来越不准确！在最坏的情况下，过滤器会发散。当我们开始研究非线性滤波器时，我们会看到很多。

卡尔曼滤波器是世界的数学模型。输出仅与该模型一样准确。为了使数学易于处理，我们必须做出一些假设。我们假设传感器和运动模型具有高斯噪声。我们假设一切都是线性的。如果这是真的，卡尔曼滤波器在最小二乘意义上是*最优的。这意味着没有办法做出比过滤器给我们的更好的估计。然而，这些假设几乎从来都不是真的，因此模型必然是有限的，工作过滤器很少是最优的。

在后面的章节中，我们将处理非线性问题。现在我想让你明白，设计线性滤波器的矩阵是一个实验过程，而不是一个数学过程。使用数学来建立初始值，但随后您需要进行实验。如果世界上有很多无法计算的噪音（风等），您可能必须使 $\mathbf Q$ 更大。如果您将其设置得太大，过滤器将无法快速响应更改。在**自适应滤波器**一章中，您将学习一些替代技术，这些技术允许您根据输入和性能实时更改滤波器设计，但现在您需要找到一组适用于这些条件的值你的过滤器会遇到。如果飞行员是学生，则杂技飞机的噪声矩阵可能会与飞行员是专家的情况不同，因为动态会完全不同。
